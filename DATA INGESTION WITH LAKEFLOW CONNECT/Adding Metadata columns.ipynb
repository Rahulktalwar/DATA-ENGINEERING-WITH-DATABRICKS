ADDING METADATA COLUMNS DURING INGESTION


```python
%sql
select *,
_metadata.file_name as source_file_name,
_metadata.file_path as file_path,
_metadata.file_modification_time as last_modified_timestamp,
DATE(current_timestamp()) as ingest_dt
from read_files("/Volumes/workspace/raw_gcd/contracts_file/future_jobs_dataset.csv")
limit 10
```

```python
%sql
---CREATE FINAL BRONZE TABLE

CREATE OR REPLACE TABLE future_jobs_bronze AS
select *,
_metadata.file_name as source_file_name,
_metadata.file_path as file_path,
_metadata.file_modification_time as last_modified_timestamp,
DATE(current_timestamp()) as ingest_dt 
from read_files("/Volumes/workspace/raw_gcd/contracts_file/future_jobs_dataset.csv")
```

```python
%sql
SELECT * FROM future_jobs_bronze 
limit 10 
```

```python
%sql
----COUNT TOTAL RECORDS OF FILE

SELECT source_file_name, count(*)
FROM future_jobs_bronze
group by source_file_name
```

**Working with the Rescued Data Column**


read_files(), spark.read or Auto Loader provides a rescued data column if the raw data doesn't match the schema.

The rescued data column ensures that columns that don't match with the schema are rescued instead of being dropped.
 The rescued data column contains any data that isn't parsed for the following reasons:

The column is missing from the schema.
Type mismatches.
Case mismatches.

```python
%sql
---READ THE CSV FILE
select * from read_files("/Volumes/workspace/raw_gcd/contracts_file/future_jobs_dataset.csv",
format=>'csv',
header=>'true'
)
limit 10  ;
```
